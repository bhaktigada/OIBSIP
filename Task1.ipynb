```python
# Importing necessary libraries for model training and evaluation
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics


# Splitting the data into training and testing sets
train, test = train_test_split(iris_data, test_size=0.75)

# Define features and target variables for training and testing sets
features = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
train_X = train[features]
train_y = train['Species']
test_X = test[features]
test_y = test['Species']

# Initialize the KNN classifier
knn_model = KNeighborsClassifier(n_neighbors=3)

# Fit the model on the training data
knn_model.fit(train_X, train_y)

# Make predictions on the test data
predictions = knn_model.predict(test_X)

# Calculate and print the accuracy
accuracy = metrics.accuracy_score(predictions, test_y)
print('KNN Accuracy:', accuracy)
```
